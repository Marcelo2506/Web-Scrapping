```python
from urllib.request import urlopen, urlretrieve, Request
from urllib.error import URLError, HTTPError
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.fundamentus.com.br/resultado.php'
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36 Edg/121.0.0.0'}
```

```python
req = Request(url,headers=headers)
response = urlopen(req)  #para abrir a url
html = response.read() #abrir o html
soup = BeautifulSoup(html,'html.parser')
```

```python
lista = soup.find('table')
```

```python

#quantidade de ações
qtd = soup.find_all('span', class_='tips')
qtd = range(int(len(qtd)-1)) #retira o ultimo registro para não ter erro
#qtd que será utilizada no for
resumo = []
```

```python
qtd
```

```python
#Pega a informações que não entram no for
papel = lista.find('td').find('span',class_='tips').getText() #Nome da ação
cotacao = lista.find('td').findNext('td').contents[0] #Cotação
```

```python
for i in qtd:

    acoes ={} 

    PL= cotacao.find_next('td').contents[0]
    PVP= PL.find_next('td').contents[0]
    PSR= PVP.find_next('td').contents[0]
    DIV_YIELD= PSR.find_next('td').contents[0]
    PATIVO= DIV_YIELD.find_next('td').contents[0]
    PCAPGIRO=PATIVO.find_next('td').contents[0]
    PEBIT= PCAP_GIRO.find_next('td').contents[0]
    PATIVCIRC_LIQ= PEBIT.find_next('td').contents[0]
    EVEBIT= PATIVCIRC_LIQ.find_next('td').contents[0]
    EVEBITDA= EVEBIT.find_next('td').contents[0]
    MRGEBIT= EVEBITDA.find_next('td').contents[0]
    MRG_LIQ= MRGEBIT.find_next('td').contents[0]
    LIQ_CORR= MRG_LIQ.find_next('td').contents[0]
    ROIC= LIQ_CORR.find_next('td').contents[0]
    ROE= ROIC.find_next('td').contents[0]
    LIQ2MESES= ROE.find_next('td').contents[0]
    PRATLIQ= LIQ_2MESES.find_next('td').contents[0]
    DIVBRUT= PRAT_LIQ.find_next('td').contents[0]
    CRESC_REC= DIV_BRUT.find_next('td').contents[0]

    acoes['id'] = 1
    acoes['papel'] = papel
    acoes['cotacao'] = cotacao
    acoes['PL'] = PL
    acoes['PVP'] = PVP
    acoes['PSR'] = PSR
    acoes['DEV_YIELD'] = DIV_YIELD
    acoes['PATIVO'] = PATIVO
    acoes['PCAP_GIRO'] = PCAPGIRO
    acoes['PEBIT'] = PEBIT
    acoes['PATIVCIRC'] = PATIVCIRC_LIQ
    acoes['EVEBIT'] = EVEBIT
    acoes['EVEBITDA'] = EVEBITDA
    acoes['MRGEBIT'] = MRGEBIT
    acoes['MRG_LIQ'] = MRG_LIQ
    acoes['LIQ_CORR'] = LIQ_CORR
    acoes['ROIC'] = ROIC
    acoes['ROE'] = ROE
    acoes['LIQ_2MESES'] = LIQ2MESES
    acoes['PRAT_LIQ'] = PRATLIQ
    acoes['DIV_BRUT'] = DIVBRUT
    acoes['CRESC_REC'] = CRESC_REC

    resumo.append(acoes)

    try:
        papel = CRESC_REC.find_next('td').span.a.contents[0]
        cotacao = papel.findPrevious('td').findNext('td').contents[0]

    except HTTPError as e:
        print(e.status, e.reason)
```

```python
dataset = pd.DataFrame(resumo)
```
